## 1. Curation Methodology

### Inclusion criteria
- The prediction must name a specific year or be clearly datable to a specific year.
- The predictor must be named (individual, institution, or government agency).
- The prediction must be about IT or AI showing up in aggregate productivity statistics (TFP, labor productivity, GDP).
- A primary or strong secondary source must be available.

### Exclusion criteria
- Deliberately vague predictions ("AI will eventually raise productivity").
- Predictions about specific firms or sectors without aggregate claims.
- Unsourced or unverifiable claims.

### Schema
Each prediction carries: predictor, predictor_type, institution, category, stance, year_said, year_targeted, quote, context, source_url, metric_predicted, magnitude_predicted, outcome, outcome_note, verification_confidence, tags.

## 2. BLS Data Pull

### Series used
- **PRS85006152**: BLS Labor Productivity, nonfarm business sector (annual % change, index 2017=100)
- **MPU4900013**: BLS Total Factor Productivity, nonfarm business sector (annual, with lag)

### Pull instructions
```python
import requests

# BLS Public Data API — free, registration optional for small volumes
url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
headers = {'Content-Type': 'application/json'}
payload = {
    'seriesid': ['PRS85006152', 'MPU4900013'],
    'startyear': '1985',
    'endyear': '2025',
    'registrationkey': 'YOUR_KEY'  # optional for < 500 calls/day
}
response = requests.post(url, json=payload, headers=headers)
data = response.json()
```

### Vintage
Data vintage: 2026-04-30. TFP through 2023 (BLS annual publication with long lags). Labor productivity through 2025.

### Refresh cadence
Quarterly, after BLS Productivity and Costs release (~40 days after quarter end).

## 3. Verification Methodology

### Confidence levels
- **High**: Direct primary source — NBER paper, FRB working paper, CBO document, official transcript, published book.
- **Medium**: Secondary source with clear citation — reputable press, well-known interview, institutional report.
- **Low**: Paraphrased, secondhand, or unclear original source. Flagged prominently in the archive.

### Target: 85%+ high confidence
Current archive: ~78% high, ~18% medium, ~4% low. The medium/low entries are primarily CEO interviews and sell-side research notes where the original document is not publicly archived.

## 4. Known Gaps

- **European predictors**: The archive is heavily US-centric. OECD and BIS entries provide some international coverage, but European central bank predictions are underrepresented.
- **Pre-1987**: Drucker (1978) and Bell (1979) are included for historical depth, but the archive focuses on the Solow Paradox era (1987+).
- **Sector-specific**: Only a handful of sector-specific predictions are included. Most aggregate predictions ignore sectoral heterogeneity.
- **Wage effects**: Only two entries (Autor 2015, 2019) explicitly address wage-productivity links. This could be expanded.
- **Live updates**: Post-2022 predictions are mostly pending. The archive will need quarterly updates as target years arrive and BLS data refreshes.
